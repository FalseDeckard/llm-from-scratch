```python
with open("the-verdict.txt", "r", encoding="utf-8") as file:
    text = file.read()

print("Total char count:", len(text))
print(text[:99])
```

```python
import re

preprocessed = re.split(r'([^\w\s]|--|\s+)', text)
preprocessed = [word.strip() for word in preprocessed if word.strip()]
print(len(preprocessed))
print(preprocessed[:99])
```

```python
all_words = sorted(set(preprocessed))
vocab_size = len(all_words)
print((vocab_size))
```

```python
vocabulary = {token:integer for integer, token in enumerate(all_words)}
for i, item in enumerate (vocabulary.items()):
    print(item)
    if i >= 99:
        break
```

```python
class SimpleTokenizerv1:
    def __init__(self, vocabulary):
        self.str_to_int = vocabulary
        self.int_to_str = {v: k for k, v in vocabulary.items()}

    def encode(self, text):
        preprocessed = re.split(r'([^\w\s]|--|\s+)', text)
        preprocessed = [word.strip() for word in preprocessed if word.strip()]
        ids = [self.str_to_int[s] for s in preprocessed]
        return ids
    
    def decode(self, ids):
        text = " ".join([self.int_to_str[i] for i in ids])

        text = re.sub(r'\s+([,.?!()\"\'])', r'\1', text)
        return text
```

```python
tokenizer = SimpleTokenizerv1(vocabulary)
text = "I looked about the spacious white-panelled room, with its _famille-verte_ vases repeating the tones of the pale damask curtains, and its eighteenth-century pastels in delicate faded frames."
encoded = tokenizer.encode(text)
decoded = tokenizer.decode(encoded)
print("Encoded:", encoded[:99])
print("Decoded:", decoded[:99])
```
